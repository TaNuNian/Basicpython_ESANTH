## **ขั้นตอนการพัฒนาตัวแบบการเรียนรู้ของเครื่อง**

 การเรียนรู้ของเครื่อง (Machine learning) นั้นเป็นการสร้างตัวแบบโดยอาศัยการเรียนรู้รูปแบบลักษณะที่สำคัญของข้อมูลเป็นหลัก โดยขั้นตอนงานสำหรับขั้นตอนการพัฒนาตัวแบบการเรียนรู้ของเครื่อง (Machine learning workflow) ประกอบด้วยลำดับขั้นของงานย่อยดังนี้.   

 <img src = 'https://drive.google.com/uc?id=1vV_d6KIZdoGah1rz-THDvXnQ_2lf6S2W' alt = 'ML workflow' width ='700px' height = '500px'>  

### **1.ดึงข้อมูล (Retrieve data)**

เป็นการเข้าถึงและดึงข้อมูลมาจากแหล่งเก็บข้อมูลหลายรูปแบบ เช่น spread sheet ฐานข้อมูล หรือ เว็บไซต์ เป็นต้น เพื่อนำมาใช้สร้างและพัฒนาตัวแบบ

### **2.สำรวจและทำความสะอาดข้อมูล (Explore and Clean data)**
เป็นการตรวจสอบลักษณะของข้อมูลซึ่งประกอบด้วย

*   ประเภทของตัวแปร
*   จำนวนข้อมูล
*   จำนวนตัวแปรแต่ละประเภท
*   จำนวนข้อมูลสูญหาย
*   ข้อมูลผิดปกติ
*   ข้อมูลเชิงสถิติของแต่ละตัวแปร เช่น min max mean std เป็นต้น

สำหรับการทำความสะอาดข้อมูล มีรายละเอียดดังนี้

*   การจัดการข้อมูลสูญหาย
*   การจัดการความไม่สอดคล้องของข้อมูล
*   การจัดการข้อมูลที่ผิดปกติ เช่น การลบข้อมูลผิดปกติ หรือการเติมค่าข้อมูลสูญหาย เป็นต้น    

### **3.เตรียมข้อมูลและการสกัดคุณลักษณะเด่น (Prepare data and Extract feature)**
หลายตัวแบบหรือวิธีการในกลุ่มการเรียนรู้ของเครื่อง (Machine learning) จำเป็นจะต้องปรับลักษณะของข้อมูลก่อนการใช้ตัวแบบเช่นการทำปรับข้อมูลชนิดข้อมูลต่อเนื่อง (Continuous)ให้เป็นข้อมูลเต็มหน่วยเพื่อลดความแปรปรวนของข้อมูล ซึ่งกระบวนการนี้จะเรียกว่าเป็นการเตรียมข้อมูลให้เหมาะสมและสอดคล้องตามเงื่อนไขหรือข้อกำหนดของการใช้ตัวแบบ

ส่วนการสกัดคุณลักษณะหรือที่เรียกว่า**ฟีเจอร์ (Feature)** ที่สำคัญจากตัวแปรตั้งต้น เป็นขั้นตอนที่สำคัญขั้นตอนหนึ่งที่ช่วยในกระบวนการสอนตัวแบบ นอกจากนี้ในขั้นตอนนี้ ยังรวมไปถึงการแบ่งข้อมูล (Data splitting)   สำหรับใช้ในการพัฒนาตัวแบบ และการทดสอบตัวแบบ ตัวอย่างเช่น แบ่งข้อมูลเป็น 2 ชุดย่อย ชุดข้อมูลแรกเรียกว่า "*ชุดข้อมูลสอน (Training dataset)*" อีกชุดข้อมูลเรียกว่า "*ชุดข้อมูลทดสอบ (Test set)*" หรือ แบ่งข้อมูลเป็น 3 ชุดย่อย ชุดข้อมูลแรกเรียกว่า "*ชุดข้อมูลสอน (Training dataset)*" ชุดข้อมูลที่สองเรียกว่า "*ชุดข้อมูลตรวจสอบ (Validate dataset)*" และ ชุดข้อมูลที่สามเรียกว่า "*ชุดข้อมูลทดสอบ (Test dataset)*" เป็นต้น  

### **4.พัฒนาและสอนตัวแบบ (Validate and train model)**
ในขั้นตอนนี้เป็นส่วนของการเลือกตัวแบบที่เหมาะสมกับปัญหา เช่นหากเป็นปัญหาที่เกี่ยวกับการจำแนกประเภทของข้อมูล ตัวแบบที่จะใช้จะอยู่ในกลุ่มตัวแบบการเรียนรู้แบบมีผู้สอน (Supervised learning) หรือหากเราต้องการหากฎความความสัมพันธ์ (Association rule) ของการซื้อสินค้า เราก็อาจจะเลือกตัวแบบที่ในกลุ่มการเรียนรู้แบบไม่มีผู้สอน (Unsupervised learning) เป็นต้น ซึ่งหลายๆครั้งเราก็จำเป็นต้องพัฒนาตัวแบบหลายๆตัวแบบโดยผ่านการเรียนรู้จากข้อมูลชุดสอน    

### **5.ตรวจสอบและประเมินตัวแบบ (Validate and Evaluate model)**
หลายตัวแบบการเรียนรู้ของเครื่องก่อนการสร้างตัวแบบเราจำเป็นที่จะต้องกำหนดค่าพารามิเตอร์ที่เกี่ยวข้องกับตัวแบบก่อนที่จำทำการสอนตัวแบบ เช่น ตัวแบบต้นไม้ตัดสินใจ (Decision tree) จะมีการกำหนดจำนวนความลึกของต้นไม้ หรือจำนวนเร็คคอร์ดที่น้อยที่สุดที่จะให้เป็นโหนดของต้นไม้ เป็นต้น ดังนั้นเราจึงต้องมีการตรวจสอบตัวแบบด้วยชุดข้อมูลตรวจสอบ (Validate dataset) เพื่อให้ได้ค่าพารามิเตอร์ที่เหมาะสมของตัวแบบ จากนั้นเราจึงทำการทดสอบประสิทธิภาพของตัวแบบด้วยชุดข้อมูลทดสอบ (Test dataset)ต่อไป  

### **6.นำใช้ตัวแบบ (Deploy model)**
หลังจากที่เราได้ตัวแบบการเรียนรู้ของเครื่องที่เหมาะสม ในขั้นตอนต่อไปก็เป็นการนำตัวแบบที่ได้ไปใช้งาน โดยอาจเป็นการใช้งานตัวแบบผ่านเว็บแอพลิเคชัน หรือ เว็บ API เป็นต้น

### **7.ติดตามและปรับปรุงตัวแบบและปรับปรุงข้อมูล (Monitor model)**
หลังจากใช้ตัวแบบไปซักระยะหนึ่งสิ่งที่เราควรทำคือการติดตามว่าตัวแบบยังทำงานได้ดีหรือไม่เนื่องจากเมื่อเวลาผ่านไปลักษณะ พฤติกรรม หรือการแจกแจงของข้อมูลอาจเกิดการเปลี่ยนแปลงไปให้ให้ประสิทธิภาพของตัวแบบลดน้อยลง ดังนั้นจึงจำเป็นที่เราจะต้องมีการปรับปรุงข้อมูลที่ใช้ในการสอน ตรวจสอบและทดสอบตัวแบบอีกครั้ง
