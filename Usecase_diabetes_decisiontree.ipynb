```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor, export_graphviz

x = df.drop(['DiabetesPedigreeFunction'],axis=1)
y = df.DiabetesPedigreeFunction
```

```python
from sklearn.model_selection import train_test_split
# Split dataset into training set and test set
train_data, test_data, train_y, test_y = train_test_split(x, y, test_size=0.20, shuffle=True)
```

```python
train_data
```

```python
# Create Decision Tree classifer object
#decision_tree_classifier = DecisionTreeClassifier(criterion="gini")
decision_tree_regressor = DecisionTreeRegressor()

# Train Decision Tree Classifer
#decision_tree_classifier.fit(train_data,train_y)
decision_tree_regressor.fit(train_data,train_y)

#y_predict = decision_tree_classifier.predict(test_data)
y_predict = decision_tree_regressor.predict(test_data)

y_predict
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(test_y,y_predict))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

print("MSE=", mean_squared_error(test_y,y_predict))
print("MAE=", mean_absolute_error(test_y,y_predict))
```
