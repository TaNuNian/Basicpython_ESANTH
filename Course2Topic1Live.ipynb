## ตัวอย่างโปรแกรมแบบที่ 1

````python
customers=[['คนที่ 1','รถยนต์',10],['คนที่ 2','รถจักรยานยนต์',40],['คนที่ 3','รถยนต์',80],['คนที่ 4','รถยนต์',3360],['คนที่ 5','รถยนต์',15],['คนที่ 6','รถยนต์',421],['คนที่ 7','รถจักรยานยนต์',16],['คนที่ 8','รถยนต์',1442],['คนที่ 9','รถยนต์',2800],['คนที่ 10','รถจักรยานยนต์',1440]]
Ncustomers=len(customers)
#print(Ncustomers)
import math
for i in range(Ncustomers):
    #print(i+1,customers[i][2])
    vehicle=customers[i][1]
    #print(vehicle)
    tminutes=customers[i][2]
    #print(tminutes)
    days=math.floor(tminutes/(60*24))
    hours=(tminutes-days*24*60)/60
    minutes=tminutes%60                  # คำนวณจำนวนนาทีส่วนที่เหลือที่ไม่ถึงชั่วโมง
    totalhours=math.floor(tminutes/60)   # คำนวณจำนวนเวลาส่วนที่เต็มชั่วโมง
    #print(days,hours)
    fee=0
    if vehicle=='รถจักรยานยนต์':
        if hours <= 0.25:
            fee=0
        elif hours <= 1:
            fee=10
        elif hours <= 2:
            fee=20
        elif hours <= 3:
            fee=30
        elif hours <= 4:
            fee=40
        elif hours <= 5:
            fee=50
        elif hours <= 6:
            fee=60
        else:
            fee=70
        fee = fee + days*70
    else: #รถยนต์
        if hours <= 0.25:
            fee=0
        elif hours <= 1:
            fee=25
        elif hours <= 2:
            fee=50
        elif hours <= 3:
            fee=80
        elif hours <= 4:
            fee=110
        elif hours <= 5:
            fee=145
        elif hours <= 6:
            fee=180
        else:
            fee=250
        fee = fee + days*250
    print('********************************************************')
    print('ใบเสร็จรับเงินค่าจอดรถ')
    print('ชื่อผู้เข้าใช้บริการลานจอดรถ: {}    ประเภทพาหนะ: {}'.format(customers[i][0],customers[i][1]))
    print('จำนวนเวลา: {:d} ชั่วโมง {:2d} นาที รวมเป็นเงิน {} บาท'.format(totalhours,minutes,fee))
    print('********************************************************')
````
